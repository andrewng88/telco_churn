```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# suppress warning
import warnings
warnings.filterwarnings("ignore")

# set diverging color palette
# https://chrisalbon.com/python/data_visualization/seaborn_color_palettes/
color = sns.color_palette('CMRmap')
sns.palplot(color)
sns.set()
```

**Problem Statement**

Improve sales revenue of the telco by predicting(classifiying) customer Churn/No Churn using the provided features.
Accuracy must be the same or better than baseline but we 'gotta-catch'em all' the Churn customers. 

```python
# read csv and display 5 rows
df = pd.read_csv('data/churn.csv')
df.head()

# drop 'customerID' as it's a serialize number
df.drop(columns = 'customerID',inplace=True)
```

```python
# Churn is our target variables, the rest are predictor variables
df.columns
```

```python
# Convert to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors="coerce")
df.isnull().sum().sort_values(ascending=False)
```

```python
#display all rows that are null
df[df.isna().any(axis=1)]
```

```python
# We will drop the 11 values since it's only 0.15% of the DF
# and we are not able to determine the Total Charges
# and the target variable is 'No Churn'
df.dropna(inplace=True)
```

```python
# verify there is no null
df.isnull().sum().sort_values(ascending=False)
```

```python
# backup processed df
df.to_csv('data/churn_done.csv',index=False)
```

There are 3 types of data:
* Numerical - continuous values - 10.1cm , 5.2 seconds etc 
* Categorical ( Nominal - there is no order i.e blue, green , orange)
* Categorical ( Ordinal - there is order i.e small, medium , large ) 

```python
binary_columns = ['Partner','Dependents','PhoneService','PaperlessBilling','Churn']
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
#label.fit(df['Partner'])

def convert_to_binary(binary_columns):
    '''convert a list of columns to binary'''
    for col in binary_columns:
        #df[col]= label.transform(df[col])
        df[col]= label.fit_transform(df[col])
        
convert_to_binary(binary_columns)
```

```python
# ['Partner','Dependents','PhoneService','PaperlessBilling','Churn'] already binarized
df.info()
```

## Partner and Dependents

```python
pd.concat([
           df.groupby(['Partner'])['Churn'].value_counts().round().to_frame(),
           df.groupby(['Partner'])['Churn'].value_counts(normalize=True).round(2).to_frame(name='%'), 
           ],
          axis=1)
```

```python
g = sns.catplot(x="Partner", hue='Churn',kind="count", data=df)
g.fig.suptitle('Churn based on hasPartner');
```

If there is no Partner, the tendency to Churn is higher. Dual income has lesser chance to churn

```python
pd.concat([
           df.groupby(['Dependents'])['Churn'].value_counts().round().to_frame(),
           df.groupby(['Dependents'])['Churn'].value_counts(normalize=True).round(2).to_frame(name='%'), 
           ],
          axis=1)
```

```python
g = sns.catplot(x="Dependents", hue='Churn',kind="count", data=df)
g.fig.suptitle('Churn based on hasDependents');
```

If there is no Dependents, the tendency to Churn is higher. It could be adults that has Dependents are more responsible.

## Multiple lines

```python
pd.concat([
           df.groupby(['MultipleLines'])['Churn'].value_counts().round().to_frame(),
           df.groupby(['MultipleLines'])['Churn'].value_counts(normalize=True).round(2).to_frame(name='%'), 
           ],
          axis=1)
```

Between the MultipleLines , the most occurence for each of the category are
